```python
move_label_x = 10
point_size = 300

for numb_cluster in range(len(models)):
    fig, axs = plt.subplots(3, 4, figsize=(25, 25))
    # fig, axs = plt.subplots(5, 4, figsize=(25, 25))
    count = 0

    for i in range(len(axs)):
        for k in range(len(axs[i])):
            if count > 8:
                break
            axs[i][k].imshow(im, extent=[min_x - 20, max_x + 30, min_y - 100, max_y + 30])

            # if count > 18:
            #     break
            # axs[i][k].imshow(im, extent=[min_x - 20, max_x + 30, min_y - 30, max_y + 30])
            axs[i][k].set_axis_off()
            neg_y_copy = neg_y.copy()
            x_copy = x.copy()
            neg_y_copy.pop(count)
            x_copy.pop(count)
            sc = axs[i][k].scatter(neg_y_copy, x_copy, c=models[keys_list[numb_cluster]][count], s=point_size,
                                   cmap='magma_r',
                                   vmin=min_all_parameters, vmax=max_all_parameters, edgecolors='black')

            cbar = plt.colorbar(sc, ax=axs[i][k])

            axs[i][k].scatter(-(y[count]), x[count], c="g", s=point_size)
            channels_copy = channels.copy()
            channels_copy.pop(count)

            for j, label in enumerate(channels_copy):
                axs[i][k].annotate(label, (neg_y_copy[j], x_copy[j] + move_label_x))
            axs[i][k].annotate(channels[count], (-(y[count]), x[count] + move_label_x))
            axs[i][k].set_title(f'Electrode:{channels[count]}')
            count += 1
    fig.delaxes(axs[len(axs) - 1][len(axs[i]) - 1])
    fig.delaxes(axs[len(axs) - 1][len(axs[i]) - 2])
    fig.delaxes(axs[len(axs) - 1][len(axs[i]) - 3])
    fig.suptitle(f'Electrodes relationship in cluster{numb_cluster + 1}', fontsize=16)
    fig.set_facecolor('white')
    plt.savefig(rf'..\reports\electrodes-relation-cl{numb_cluster + 1}')
```
