```python
%matplotlib inline
import mne
import pandas as pd

data = pd.read_csv('../datasets/depression_patients.csv')
channels = ['Fp1',
            'Fp2',
            'F3',
            'F4',
            'C3',
            'C4',
            'P3',
            'P4',
            'O1',
            'O2',
            'F7',
            'F8',
            'T3',
            'T4',
            'T5',
            'T6',
            'Fz',
            'Cz',
            'Pz', ]
n_channels = len(channels)
sfreq = 250
```

```python
info = mne.create_info(

    ch_names = channels,
    ch_types = ['eeg']*n_channels,
    sfreq=sfreq
)
info.set_montage('standard_1020')
print('Event created:', info)
```
